#### **Exercici 1**

Definim la funció

```python
def my_function(u):
    w=2.0*np.pi
    f=np.sin(w*u)
    return f
```

Creem el domini de la funció

```python
x_list = np.arange(0.0, 2.0, 0.01)
num_punts=len(x_list)
print(num_punts)
```

Avaluem la funció

```python
y=my_function(x_list)
print(y)
```

Creem matrius buides per a la primera derivada (x i y)

```python
x1=[]
y1=[]
```

```python
d=0
while d <num_punts-1:
     derivative=(y[d+1]-y[d])/(x_list[d+1]-x_list[d])
     y1.append(derivative)
     x1.append(x_list[d])
     d=d+1
print(d)
```

Creem matrius buides per a la segona derivada (x i y) i fem la segona derivada

```python
x2=[]
y2=[]
i=1
while i<num_punts-1:
     second_derivative=(y[i+1]-2*y[i]+y[i-1])/((x_list[i+1]-x_list[i])**2)
     y2.append(second_derivative)
     x2.append(x_list[i])
     i=i+1
print(i)
```

Ara posem la segonda derivada de manera explicita, aixì podrem comprobar la exactitud

```python
w = 2.0 * np.pi
F_second_exact = -(w**2) * np.sin(w * x_list)
print(F_second_exact)
```

Finalment comparem els gràfics per veure si coincideixen

```python
plt.plot(x2, y2, 'r', label='Numerical Second Derivative')
plt.plot(x_list[1:num_punts-1], F_second_exact[1:num_punts-1], 'b--', label='Exact Second Derivative')
plt.title('Numerical and Exact Second Derivatives')
plt.xlabel('x')
plt.ylabel('Second Derivative')
plt.legend()
plt.show()
```

#### **Exercici 2**

El primer que farem serà definir el temps i els parametres fisics

```python
m=1
k=1
t_inici=0
t_final=20
dt=0.01
t_llista=np.arange(t_inici, t_final, dt)
num_punts= len(t_llista)
```

Definim les condicions inicials de com estara la motlla

```python
x0 = 1.0       # posició inicial
v0 = 0.0       # velocitat inicial
```

Creem llistes buides per almatzenar els valors

```python
x_num= []
v_num = []
a_num= []
```

Guardem els primers valors

```python
x0=x0
a0= -(k/m) * x0
print (a0)
print (x0)
x_num.append(x0)
v_num.append(v0)
a_num.append(a0)
```

Necesitem dos punts per l'equació de Verlet pugui "predir el futur". Per tant calculem el punt 1 ja que ja tenim el punt x0

```python
x1 = x0 + v0*dt + 0.5*a0*dt**2
x_num.append(x1)
```

Ara apliquem el algoritme de Verlet

```python
i=1
while i<num_punts-1:
     ai=-(k/m) * x_num[i]
     xi= 2*x_num[i] - x_num[i-1] + ai*dt**2
     x_num.append(xi)
     i=i+1
print(i)
```

Fem la representació gràfica.

```python
plt.plot (t_llista, x_num, "g")
plt.xlabel("Temps (t)")
plt.ylabel("Posició (x(t))")
plt.title("Oscil·lador harmònic - Posició (Verlet)")
plt.show()
```

#### **Exercici 3**

El primer que fem es definir els parametres.

```python
N = 200
L = 1.0
h = L / (N + 1)
mu = 1.0
x = np.linspace(0, L, N)
print (h)
```

Definim el potencial del pou infinit dins del pou, que serà 0.

```python
V = np.zeros(N)
```

Començem creant la matriu

```python
F = np.zeros((N, N))
```

```python
i=1
for i in range(N):
    F[i, i] = -2/mu - 2*h**2 * V[i]
    if i > 0:
        F[i, i-1] = 1/mu
    if i < N-1:
        F[i, i+1] = 1/mu
```

Comencem a resoldre, o farem diagonalitzant la matriu

```python
w, psi = np.linalg.eigh(F)
print (w,psi)
```

Calcularem els nivells de energia i ens fixarem en els primers

```python
E = -w / (2*h**2)
print(E[:7])
```

grafiquem la funció d'ona corresponent a l'estat fonamental

```python
plt.plot(x, psi[:, 0])
plt.xlabel("x")
plt.ylabel("Ψ(x)")
plt.show()
```
