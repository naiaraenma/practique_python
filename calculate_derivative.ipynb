## Numerical evaluation of the derivative of a function from definition

Import python libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definir una funció

U es un valor arbitrari, no es un valor. Després quan calculem y posem (...) aquest sera el valor

```python
def my_function(u):
    w=2.0*np.pi
    f=np.sin(w*u)
    return f
```

Create the domain of the function

```python
x_list = np.arange(0.0, 2.0, 0.001)
num_punts=len(x_list)
print(num_punts)
```

Evaluate the function (example: sin(w x))

```python
y=my_function(x_list)
print(y)
```

Create empty arrays for the derivative (x and y)

```python
x2=[]
y2=[]
```

Loop over these points

append es per afegir els punts calculats en la llista buida que hem creat anteriorment. Important que ha d¡haver-hi l'espai perque es fasi el loop, sino no ho farà.

```python
i=0
while i<num_punts-1:
     derivative=(y[i+1]-y[i])/(x_list[i+1]-x_list[i])
     y2.append(derivative)
     x2.append(x_list[i])
     i=i+1
print(i)
```

Plot the results

```python
plt.plot(x_list, y, 'r', x2, y2, 'b')
plt.xlabel('x')
plt.ylabel('f and df/dx')
plt.title('Numerical calculation of derivative')
plt.legend(['f(x)', 'df/dx'])
plt.show()
```
