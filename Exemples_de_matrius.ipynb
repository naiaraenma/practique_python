## Exemples Algebra

```python
import numpy as np #importem la llibreria
```

#### Exemples de vectors (array)

```python
v=np.array([1,0,0])
u=np.array([0,-2,0])
```

```python
print (u)
print (v)
```

Operacións amb vectors (suma, resta, producte escalar i vectorial)

```python
w1=u+v #suma vectors
print(w1)
w2=u-v #resta vectors
print(w2)
```

```python
a=np.dot(u,v) #producte escalar
print(a)
```

```python
b=np.cross(u,v) #producte vectorial
print(b)
```

#### Exemples amb Matrius

```python
A=np.array([[2,4],[1,3]])
print (A)
```

```python
e=A[0,1] #si volem accedir a cada element es fa d'aquesta manera (el 0 es la primera fila i l'1 es la segona columna)
print(e)
```

v=A[0,:] #els dos punts es un comodi, no li dono index a aquest segon valor, vol dir que agafi totes les columnes de la fila 0
print (v)

```python
u=A[:,1] #li estem dient que agafi totes les files de la segona columna
print(u)
```

Valors i vectors propis

```python
A=np.array([[2,2],[1,3]])
print (A)
```

```python
Valors,Vectors=np.linalg.eig(A) #Amb això calculem valors i vectors propis, pero sense ordenar.
print("Valors propis",Valors)
print("Vectors propis",Vectors)
```

```python
u=Vectors[:,0] #Per treure el primer vector propi fem això, ja que no el volem en format de matriu
print(u)
```

```python
v=Vectors[:,1] #Així s'aconsegueix el segon vector propi
print(v)
```

Comprovem que efectivament son vectors propis

```python
z=np.dot(A,u)#La mateixa variant per multiplicar vectors serveix per multiplicar matrius
print(z)
```

```python
z2=np.dot(A,v)
print(z2)
```
